```python
import os
import sys
sys.path.insert(0, r'/home/charmi533/Desktop/Btp/code/GetOldTweets-python')
import Exporter
import shlex
import pandas as pd
import datetime
```

```python
def write_file(sentiment, start_date, end_date):
    emoticon = ''
    if sentiment == 'pos':
        emoticon = ':)'
    else:
        emoticon = ':('
    
    outputfile = 'demofile_{0}.csv'.format(sentiment)
    maxTweets = 3000
    cmd = '--querysearch "#demonetisation OR #demonetization {0}" --since {1} --until {2} --toptweets --maxtweets {3} --outputfile {4}'.format(emoticon, start_date, end_date, maxTweets, outputfile)

    Exporter.main(shlex.split(cmd))
    
```

```python
def label_tweets():
    df_pos = pd.read_csv('demofile_pos.csv', delimiter='\t', quotechar='"')
    df_neg = pd.read_csv('demofile_neg.csv', delimiter='\t', quotechar='"')
    
    len_pos = len(df_pos['text'])
    len_neg = len(df_neg['text'])
    
    text_pos = pd.DataFrame({"text": df_pos['text'], "label": [1 for i in range(len_pos)]})
    text_neg = pd.DataFrame({"text": df_neg['text'], "label": [-1 for i in range(len_neg)]})
    
    df_labeled = pd.DataFrame()
    df_labeled = df_labeled.append(text_pos)
    df_labeled = df_labeled.append(text_neg)
    
    df_labeled.to_csv('labeled_tweets.csv', sep='\t', quotechar='"', encoding='utf-8')
```

```python
def daterange(start_date, end_date):
    for n in range((end_date - start_date).days + 1):
        yield start_date + datetime.timedelta(n)
```

```python
def main():
    start_date = datetime.date(year=2016, month=12, day=16)
    end_date = datetime.date(year=2016, month=12, day=29)
    for date in daterange(start_date, end_date):
        write_file('pos', str(date), str(date+datetime.timedelta(1)))
        write_file('neg', str(date), str(date+datetime.timedelta(1)))
    label_tweets()
```

```python
if __name__ == '__main__':
    main()
```
